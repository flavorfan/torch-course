## CUDA and GPUs Explained

GPUs are most suitable for parallel computing. This is why GPUs are heavily used deep neural networks. Neural networks are said to be __embarassingly parallel__, that is, can easily be trivially broken into computations that are independent of each other. 

CUDA is a software platform that provides an API for developers for interacting with the GPU. For example `cuDNN` is the CUDA deep learning neural network library. For PyTorch, CUDA is integrated into the PyTorch, one does not know how to interact with the CUDA Toolkit directly. At bottleneck points, PyTorch moves to C/C++ and CUDA.
